```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img = plt.imread('IMG_20201208_082134.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
type(img)
```

```python
print(img.shape)
print(img.size)
```

```python
w,h,d = img.shape
image_array =  img.reshape(w*h,d)
print(image_array.shape)
image_array=image_array/255
```

```python
from sklearn.utils import shuffle
image_array_sample = shuffle(image_array,random_state=10)[:5000]
image_array_sample.size
```

```python
km=KMeans(n_clusters=20,random_state=10)
km.fit(image_array_sample)
```

```python
labels=km.predict(image_array)
labels
```

```python
print(km.cluster_centers_)
c =km.cluster_centers_
```

```python
def recreate_image(c,labels,w,h,d):
    image=np.zeros((w,h,d))
    
    label_idx=0

    for i in range (w):
        for j in range (h):
            image[i][j]=c[labels[label_idx]]
            label_idx+=1
    return(image)
    
        

        
```

```python
plt.figure(1)
plt.axis('off')
plt.title('original')
plt.imshow(img)
plt.show()

plt.figure(2)
plt.axis('off')
plt.title('reduced')
plt.imshow(recreate_image(c,labels,w,h,d))
plt.show()
```
